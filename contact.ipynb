```python
contacts = {}

def add_contact(name, phone_number, email):
    contacts[name] = {'phone': phone_number, 'email': email}
    print(f"Contact '{name}' added successfully.")

def view_contact(name):
    if name in contacts:
        print(f"Name: {name}")
        print(f"Phone: {contacts[name]['phone']}")
        print(f"Email: {contacts[name]['email']}")
    else:
        print(f"Contact '{name}' not found.")

def edit_contact(name, phone_number=None, email=None):
    if name in contacts:
        if phone_number:
            contacts[name]['phone'] = phone_number
        if email:
            contacts[name]['email'] = email
        print(f"Contact '{name}' edited successfully.")
    else:
        print(f"Contact '{name}' not found.")

def delete_contact(name):
    if name in contacts:
        del contacts[name]
        print(f"Contact '{name}' deleted successfully.")
    else:
        print(f"Contact '{name}' not found.")

def list_contacts():
    if contacts:
        print("List of contacts:")
        for name in contacts:
            print(name)
    else:
        print("No contacts found.")

while True:
    print("\nMenu:")
    print("1. Add Contact")
    print("2. View Contact")
    print("3. Edit Contact")
    print("4. Delete Contact")
    print("5. List All Contacts")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        name = input("Enter name: ")
        phone = input("Enter phone number: ")
        email = input("Enter email: ")
        add_contact(name, phone, email)
    elif choice == '2':
        name = input("Enter name to view: ")
        view_contact(name)
    elif choice == '3':
        name = input("Enter name to edit: ")
        phone = input("Enter new phone number (press Enter to keep unchanged): ")
        email = input("Enter new email (press Enter to keep unchanged): ")
        edit_contact(name, phone, email)
    elif choice == '4':
        name = input("Enter name to delete: ")
        delete_contact(name)
    elif choice == '5':
        list_contacts()
    elif choice == '6':
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number from 1 to 6.")
```
